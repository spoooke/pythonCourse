```python
import numpy as np
import pandas as pd

# traitement réalisés précédemment
prets = pd.read_csv('/Users/sebastiengillio/Desktop/Perso/CODE/Python/Python Pandas/manipulezUnDataframe - dataJoint/raw.githubusercontent.com_OpenClassrooms-Student-Center_fr-4452741-decouvrez-les-librairies-python-pour-la-data-science_main_data_prets.csv')

# calcul du taux d'endettement
prets['taux_endettement'] = round(prets['remboursement'] * 100 / prets['revenu'], 2)

# renommer taux en taux_interet
prets.rename(columns={'taux':'taux_interet'}, inplace=True)

# calculer le cout total du pret
prets['cout_total'] = prets['remboursement'] * prets['duree']

# calculer les bénéfices mensuels réalisés
prets['benefices'] = round((prets['cout_total'] * prets['taux_interet']/100)/(24), 2)

prets.head()

# Identifier les personnes ayant dépassé le taux d'endettement autorisé de 35 %
clients_risque = prets[prets['taux_endettement'] > 35]
nombre_de_clients_risque = len(clients_risque)

# Afficher le nombre de clients à risque
print("Nombre de clients à risque : ", nombre_de_clients_risque)

# Afficher les détails des clients à risque (clients_risque)
print(clients_risque)


# Identifier les personnes en surendettement pour l'agence parisienne
clients_risque_paris = prets[(prets['ville'] == 'PARIS') & (prets['taux_endettement'] > 35)]
nombre_de_clients_risque_paris = len(clients_risque_paris)

# Afficher le nombre de clients à risque à Paris
print("Nombre de clients à risque à Paris : ", nombre_de_clients_risque_paris)

# Afficher les détails des clients à risque à Paris (clients_risque_paris)
print(clients_risque_paris)


# Créer une nouvelle colonne "risque" avec une valeur par défaut de "Non"
prets['risque'] = "Non"

# Utiliser la méthode apply pour modifier les lignes appropriées
prets['risque'] = prets.apply(lambda row: "Oui" if row['taux_endettement'] > 35 else "Non", axis=1)


# Compter les prêts automobiles
nombre_de_prets_automobiles = len(prets[prets['type'] == 'automobile'])
print("Nombre de prêts automobiles accordés : ", nombre_de_prets_automobiles)

# Calculer le coût total moyen des prêts automobiles
cout_total_prets_automobiles = prets[prets['type'] == 'automobile']['cout_total'].sum()
cout_moyen_prets_automobiles = cout_total_prets_automobiles / nombre_de_prets_automobiles
print("Coût total moyen des prêts automobiles : ", cout_moyen_prets_automobiles)


```

```python
# Filtrer les prêts accordés par l'agence toulousaine
prets_agence_toulouse = prets[prets['ville'] == 'TOULOUSE']

# Calculer le bénéfice mensuel total réalisé par l'agence toulousaine
bénéfice_mensuel_total_toulouse = prets_agence_toulouse['benefices'].sum()
print("Bénéfice mensuel total réalisé par l'agence de Toulouse : ", bénéfice_mensuel_total_toulouse)
```
